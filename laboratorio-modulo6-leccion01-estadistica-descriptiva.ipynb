```python
datos = [15, 21, 25, 32, 35, 45, 50, 50, 55, 60]

def media (datos:list): 
    return sum(datos)/len(datos)


def mediana(datos:list):
    if len(datos) % 2 == 0:
        return datos[(len(datos)//2)]
    else:
        return (len(datos)/2 + (len(datos)/2)-1)/2

def moda(datos:list): 
    return max(set(datos), key=datos.count)


mediana(datos)
```

**Horas trabajadas**:

**Ventas realizadas**:
   - A cada empleado también se le hizo un seguimiento de cuántas ventas realizó. Aunque las ventas están influenciadas por las horas trabajadas, hay cierta **variabilidad**: algunos empleados pueden ser más productivos y vender más en menos tiempo, mientras que otros necesitan trabajar más para alcanzar los mismos resultados.


Resuelve los siguientes puntos:


2. **Comparación de medidas de centralización**:
   - ¿Cómo se diferencian la media y la mediana para los datos de **Ventas Realizadas**? ¿Qué indica esta diferencia sobre la distribución de las ventas?


```python

sns.histplot (data = df, x = "Ventas Realizadas", binwidth=3)
media = np.mean(df["Ventas Realizadas"])
mediana = np.median(df["Ventas Realizadas"])

plt.axvline(media, color = "red", label = "Media")

plt.axvline(mediana, color = "purple", label = "Mediana")
```

5. **Cálculo de la asimetría**:
   - Usa la función adecuada en Python para calcular la **asimetría** de los datos de **Horas Trabajadas** y **Ventas Realizadas**. 
   - Interpreta si el conjunto de datos es **simétrico**, **sesgado a la izquierda** o **sesgado a la derecha**.

```python
sns.histplot (data = df, x = "Ventas Realizadas", binwidth=3, color = "green")
skewv = skew(df["Ventas Realizadas"])


sns.histplot (data = df, x = "Horas Trabajadas", binwidth=3)
skewh = skew(df["Horas Trabajadas"])

print(f"La columna Ventas realizadas tiene una asimetría de {skewv}")

print(f"La columna Horas Trabajadas tiene una desviación de {skewh}")

print("los datos de Horas Trabajadas están sesgados a izq, mientras que con Ventas pasa al contrario")
```

6. **Cálculo de la curtosis**:

   - Explica qué implica esto sobre la forma de la distribución.

```python
sns.histplot (data = df, x = "Ventas Realizadas", binwidth=1, color = "green")
kurtv = kurtosis(df["Ventas Realizadas"])


sns.histplot (data = df, x = "Horas Trabajadas", binwidth=1)
kurth = kurtosis(df["Horas Trabajadas"])

print(f"La columna Ventas realizadas tiene una kurtosis de {kurtv}, positiva")

print(f"La columna Horas Trabajadas tiene una kurtosis de {kurth},negativa")

print("los datos de Horas Trabajadas están más distribuidos y son más planos que los de horas")
```
